```python
import numpy as np
import pandas as pd
import string
```

```python
lines = open("testput.txt").read().splitlines()
lines
```

```python
def FillDataFrame(dataFrame, dictionary):
    return dataFrame.append(dictionary, ignore_index=True)
```

```python
columns = list(string.ascii_lowercase)
print(columns)
```

```python
customDeclarationData = pd.DataFrame(columns=columns)

## Read from file and fill a row of data frame for each passport
dictionary = {}

with open("input.txt") as openFileObject:
    for line in openFileObject:
        if(line == '\n'):
            customDeclarationData = FillDataFrame(customDeclarationData, dictionary)
            dictionary.clear()
            continue
        
        for character in line.rstrip():
            dictionary[character] = True
    customDeclarationData = FillDataFrame(customDeclarationData, dictionary)
    dictionary.clear()

customDeclarationData = customDeclarationData.fillna(False)
customDeclarationData
```

```python
customDeclarationData.values.sum()
```

```python
def ReviseDictionaryEveryone(dictionary, numberOfPeople):
    for key in dictionary:
        if int(np.floor(dictionary[key] / numberOfPeople)):
            dictionary[key] = True
        else:
            dictionary[key] = False
    
    return dictionary
```

```python
customDeclarationData_Everyone = pd.DataFrame(columns=columns)

## Read from file and fill a row of data frame for each passport
dictionary_Everyone = {}
numberOfPeople = 0

with open("input.txt") as openFileObject:
    for line in openFileObject:
        if(line == '\n'):
            dictionary_Everyone = ReviseDictionaryEveryone(dictionary_Everyone,numberOfPeople)
            customDeclarationData_Everyone = FillDataFrame(customDeclarationData_Everyone, dictionary_Everyone)
            dictionary_Everyone.clear()
            numberOfPeople = 0
            continue
        
        for character in line.rstrip():
            if character in dictionary_Everyone:
                dictionary_Everyone[character] += 1
            else:
                dictionary_Everyone[character] = 1
    
        numberOfPeople += 1
    dictionary_Everyone = ReviseDictionaryEveryone(dictionary_Everyone,numberOfPeople)
    customDeclarationData_Everyone = FillDataFrame(customDeclarationData_Everyone, dictionary_Everyone)
    dictionary_Everyone.clear()
    numberOfPeople = 0

customDeclarationData_Everyone = customDeclarationData_Everyone.fillna(False)
customDeclarationData_Everyone
```
