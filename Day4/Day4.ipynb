```python
import numpy as np
import pandas as pd
import re
```

```python
lines = open("input.txt").read().splitlines()
```

```python
def FillDataFrame(dataFrame, dictionary):
    return dataFrame.append(dictionary, ignore_index=True)
```

```python
columns = ["pid", "byr", "iyr", "eyr", "hgt", "ecl", "cid"]
passportData = pd.DataFrame(columns=columns)
passportData

## Read from file and fill a row of data frame for each passport
dictionary = {}

with open("input.txt") as openFileObject:
    for line in openFileObject:
        if(line == '\n'):
            passportData = FillDataFrame(passportData, dictionary)
            dictionary.clear()
            continue
        
        for element in line.rstrip().split(" "):
            twoValues = element.split(":")
            dictionary[twoValues[0]] = twoValues[1]
    passportData = FillDataFrame(passportData, dictionary)
    dictionary.clear()

passportData
```

```python
validPassportData = passportData.loc[passportData[['pid', 'byr', 'iyr','eyr','hgt','ecl','hcl']].notnull().all(1)]
len(validPassportData.index)
```

```python
def CheckIfByrIsValid(byr):
    if ((len(str(byr)) == 4) and (int(byr) >= 1920) and (int(byr) <=2002)):
        return True
    else:
        return False

def CheckIfIyrIsValid(iyr):
    if ((len(str(iyr)) == 4) and (int(iyr) >= 2010) and (int(iyr) <=2020)):
        return True
    else:
        return False

def CheckIfEyrIsValid(eyr):
    if ((len(str(eyr)) == 4) and (int(eyr) >= 2020) and (int(eyr) <=2030)):
        return True
    else:
        return False

def CheckIfHgtIsValid(hgt):
    if((str(hgt)[-2] == 'c') and (str(hgt)[-1] == 'm') and (float(str(hgt)[:-2]) >= 150) and (float(str(hgt)[:-2]) <= 193)):
            return True
    elif((str(hgt)[-2] == 'i') and (str(hgt)[-1] == 'n') and (float(str(hgt)[:-2]) >= 59) and (float(str(hgt)[:-2]) <= 76)):
            return True
    else:
            return False

def CheckIfHclIsValid(hcl):
    ## Yes, it's a regexp. Almost built it myself too!
    match = re.search(r'^#(?:[0-9a-f]{6})$', hcl)
    if match:
        return True
    else:
        return False

def CheckIfEclIsValid(ecl):
    colorList = ['amb','blu','brn','gry','grn','hzl','oth']
    if str(ecl) in colorList:
        return True
    else:
        return False

def CheckIfPidIsValid(pid):
    if len(str(pid)) == 9:
        return True
    else:
        return False
```

```python
ValidByrPD = validPassportData[validPassportData['byr'].apply(CheckIfByrIsValid)]
ValidIyrPD = ValidByrPD[ValidByrPD['iyr'].apply(CheckIfIyrIsValid)]
ValidEyrPD = ValidIyrPD[ValidIyrPD['eyr'].apply(CheckIfEyrIsValid)]
ValidHgtPD = ValidEyrPD[ValidEyrPD['hgt'].apply(CheckIfHgtIsValid)]
ValidHclPD = ValidHgtPD[ValidHgtPD['hcl'].apply(CheckIfHclIsValid)]
ValidEclPD = ValidHclPD[ValidHclPD['ecl'].apply(CheckIfEclIsValid)]
ValidPidPD = ValidEclPD[ValidEclPD['pid'].apply(CheckIfPidIsValid)]

len(ValidPidPD.index)
```
